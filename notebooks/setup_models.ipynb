```python
###Mount Google Drive
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Create directories for models
mistral_dir = "/content/drive/MyDrive/models/mistral-7b"
e5_dir = "/content/drive/MyDrive/models/e5-base-v2"

os.makedirs(mistral_dir, exist_ok=True)
os.makedirs(e5_dir, exist_ok=True)

print("✅ Drive mounted and folders ready.")
```

```python
### Download Mistral 7B (GGUF) to Drive
from huggingface_hub import hf_hub_download

mistral_path = hf_hub_download(
    repo_id="TheBloke/Mistral-7B-Instruct-v0.1-GGUF",
    filename="mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    local_dir=mistral_dir
)

print("✅ Mistral 7B model saved at:", mistral_path)

```

```python
### Download E5-base-v2 Embedding Model to Drive
from huggingface_hub import snapshot_download

e5_path = snapshot_download(
    repo_id="intfloat/e5-base-v2",
    local_dir=e5_dir
)

print("✅ E5-base-v2 embedding model saved at:", e5_path)
```

```python
### Copy models to Colab runtime for faster access
### You can load models directly from Drive, but it’s slower.
### Better to copy once per session:

!mkdir -p /content/models/mistral-7b /content/models/e5-base-v2
!cp -r /content/drive/MyDrive/models/mistral-7b/* /content/models/mistral-7b/
!cp -r /content/drive/MyDrive/models/e5-base-v2/* /content/models/e5-base-v2/
```

```python
### Test both models
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer("/content/models/e5-base-v2")
emb = embed_model.encode("Retrieval Augmented Generation combines search with generation.")
print("✅ Embedding shape:", emb.shape)


from llama_cpp import Llama

llm = Llama(model_path="/content/models/mistral-7b/mistral-7b-instruct-v0.1.Q4_K_M.gguf", verbose=False)
response = llm("Question: What is Retrieval Augmented Generation?\nAnswer:")
print("✅ Model says:", response["choices"][0]["text"].strip())
```
